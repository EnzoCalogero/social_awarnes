```python
df_1=dataset[dataset["Question"] == "... for at least one specific reason" ].copy()
```

```python
df_1['media']=df_1['Value']
df_1.head()
```

```python
d_1_aggregated=df_1.groupby('Country').agg({'Value':  'max'})
d_1_aggregated.head(2)
```

```python
d_1_aggregated['Value']['max]'.head()
```

```python
import plotly.express as px
df_1=df_1.sort_values(by=['Value'], ascending=False)
fig = px.box(df_1, x="Country",y="Value")
fig.update_layout(
    title={
        'text': "Plot Title",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=10))
#fig.update_layout(yaxis=dict(autorange="reversed"))
fig.show()
```
